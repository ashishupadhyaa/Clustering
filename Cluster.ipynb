```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

```python
x, y = make_blobs(cluster_std=2, centers=3)
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.show()
```

```python
kmean_score = list()
sil_score = list()

for n in range(2, 15):
    kmean = KMeans(n_clusters=n)
    labels = kmean.fit_predict(x)
    kmean_score.append(kmean.inertia_)
    sil_score.append(silhouette_score(x, labels))
```

```python
plt.plot(range(2, 15), kmean_score)
plt.xlabel('Number of clusters')
plt.ylabel('Kmeans Score')
plt.show()
```

```python
plt.plot(range(2, 15), sil_score)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
k13 = KMeans(n_clusters=13)
label_13 = k13.fit_predict(x)
u_labels = np.unique(label_13)
```

```python
for i in u_labels:
    plt.scatter(x[label_13 == i , 0] , x[label_13 == i , 1])
plt.scatter(k13.cluster_centers_[:, 0], k13.cluster_centers_[:, 1], s=50, c='yellow')
labl = u_labels.tolist()
labl.append('centroid')
plt.legend(labl, title='title', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.show()
```
